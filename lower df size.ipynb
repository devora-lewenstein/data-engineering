```python
import pandas as pd
import numpy as np
```

```python
rest = pd.read_csv("restaurant2.csv")
```

```python
rest.info()
```

```python
rest.info(memory_usage='deep')
```

```python
numeric = rest.select_dtypes(include=['int64', 'float64']).drop(['latitude_x','longitude_x', 'commission'], axis=1)
```

```python
numeric.isnull().sum()
```

```python
for col in numeric.columns:
    rest[col] = rest[col].astype('int')
    rest[col] = pd.to_numeric(rest[col], downcast='integer')
```

```python
rest.info(memory_usage='deep')
```

```python
rest["gender"]=rest["gender"].astype("category")
```

```python
rest.info(memory_usage='deep')
```

```python
rest.select_dtypes(include=['object']).info()
```

```python
rest.select_dtypes(include=['object']).head()
```

```python
rest['created_at_x'] = pd.to_datetime(rest['created_at_x'])
rest['updated_at_x'] = pd.to_datetime(rest['updated_at_x'])
rest['created_at_y'] = pd.to_datetime(rest['created_at_y'])
rest['updated_at_y'] = pd.to_datetime(rest['updated_at_y'])
```

```python
rest.info(memory_usage='deep')
```
